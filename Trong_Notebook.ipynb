```python
# Dependencies
import pandas as pd
import numpy as np
import datetime as dt
```

**YouTube Analysis**

```python
# Load in youtube_csv and cat_id_csv for analysis
youtube_csv = ('Resources/USvideos.csv')
cat_id_csv = ('Resources/YTCategory.csv')
```

```python
# convert youtube_csv into a df
# .tail to view the last date in our dataframe
youtube_df = pd.read_csv(youtube_csv)
youtube_df.tail()
```

```python
# convert publish_time column into date format and remove time
youtube_df['publish_time'] = pd.to_datetime(youtube_df['publish_time']).dt.date
youtube_df.head()
```

```python
# convert cat_id_csv into a df
cat_id_df = pd.read_csv(cat_id_csv)
cat_id_df.head()
```

```python
# merge youtube_df and cat_id_df with an inner join on 'category id'
youtube_data = pd.merge(youtube_df, cat_id_df, on='category_id')
youtube_data.head()
```

```python
# clean youtube_data to filter out columns that won't contribute to our anlaysis
youtube_data_clean = youtube_data[['category_id', 'Cat_Name', 'trending_date', 'title', 'channel_title',  'tags', 'views', 'likes', 'comment_count']]
youtube_data_clean.head()
```

```python
# renamed columns for reference purposes and relatability with audience
youtube_data_clean.columns = ['Category_ID', 'Category', 'Trending_Date', 'Video_Title', 'Channel',
       'Tags', 'Total_Views', 'Total_Likes', 'Comments_Count']
youtube_data_clean.head()
```
