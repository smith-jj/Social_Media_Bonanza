```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file

Articles_Apr = "Resources/ArticlesApril2018.csv"
Articles_Feb ="Resources/ArticlesFeb2018.csv"
Articles_Jan = "Resources/ArticlesJan2018.csv"
Articles_Mar = "Resources/ArticlesMarch2018.csv"
```

```python
# Rread the CSV in pandas

df_Articles_Apr = pd.read_csv(Articles_Apr)
df_Articles_Feb = pd.read_csv(Articles_Feb)
df_Articles_Jan = pd.read_csv(Articles_Jan)
df_Articles_Mar = pd.read_csv(Articles_Mar)
```

```python
# Preview of the DataFrame
df_Articles_Apr
```

```python
# Delete extraneous column
df_Articles_Apr.columns
```

```python
# Delete extraneous column
df_Articles_Apr_clean = df_Articles_Apr[['articleID', 'headline',
       'keywords',  'newDesk',  'pubDate',
       'sectionName', 'snippet', 'typeOfMaterial', 'webURL']]
df_Articles_Feb_clean = df_Articles_Feb[['articleID', 'headline',
       'keywords',  'newDesk', 'pubDate',
       'sectionName', 'snippet', 'typeOfMaterial', 'webURL']]
df_Articles_Jan_clean = df_Articles_Jan[['articleID', 'headline',
       'keywords',  'newDesk',  'pubDate',
       'sectionName', 'snippet', 'typeOfMaterial', 'webURL']]
df_Articles_Mar_clean = df_Articles_Mar[['articleID', 'headline',
       'keywords',  'newDesk', 'pubDate',
       'sectionName', 'snippet', 'typeOfMaterial', 'webURL']]
```

```python
# Appending the dataframes togehter
df_Articles_Clean = df_Articles_Jan_clean.append(df_Articles_Feb_clean, ignore_index=True)
df_Articles_Clean = df_Articles_Clean.append(df_Articles_Mar_clean, ignore_index=True)
df_Articles_Clean = df_Articles_Clean.append(df_Articles_Apr_clean, ignore_index=True)

```

```python
# Verify dropped rows
df_Articles_Clean.columns = ['Article_ID', 'Headline', 'Keywords', 'News_Category', 'Pub_Date',
       'Section_Name', 'Snippet', 'Media_Type', 'URL']
df_Articles_Clean.count()
```

```python
df_Articles_Clean = df_Articles_Clean[df_Articles_Clean.News_Category != 'Video']
df_Articles_Clean = df_Articles_Clean[df_Articles_Clean.News_Category != 'Unknown']
df_Articles_Clean.count()
```

```python
df_Articles_Clean.groupby(['News_Category']).count()
```

```python
df_Articles_Clean['News_Category'] = df_Articles_Clean['News_Category'].replace(
   {'Arts&Leisure': 'Entertainment', 'BookReview': 'Entertainment', 'Dining': 'Entertainment'})
```

```python
df_Articles_Clean.groupby(['News_Category']).count()
```
